### Ejercicio 4

Contad el número de lugares que aparecen más de una vez en _dataframe_ (campo _location_), utilizando las funciones de la librería _pandas_. ¿Existe algún lugar donde se haya realizado más de una batalla? Comprobad qué bandos estaban implicados.

```python
# Respuesta

subset4 = df.groupby(['location']).size().to_frame(name = 'count_location').reset_index()
subset4

m4 = df.merge(subset4, how= 'left')\
            .filter(items=['location','count_location','attacker_1','defender_1'])\
            .sort_values(by=['location'])
m4f = m4[m4['count_location']>1]
m4f
```
